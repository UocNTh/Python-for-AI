# Bài tập Data Preprocessing

Download dữ liệu đính kèm và thực hiện các yêu cầu sau:

1. Hiển thị toàn bộ dữ liệu
2. Điền các giá trị bị thiếu của dữ liệu theo phương pháp trung bình với các trường có giá trị số và theo phương pháp tần số cao nhất với giá trị rời rạc
3. Hiển thị các bản ghi có sepal.length > 5 và sepal.width > 3
4. Chuẩn hóa dữ liệu theo min max
5. Liên tục hóa thuộc tính variety
6. Rời rạc hóa thuộc tính sepal.length 
7. Lấy mẫu ngẫu nhiên có lặp lại 50% dữ liệu

Đọc và hiển thị dữ liệu


```python
import pandas as pd 
import numpy as np 
```

```python
data = pd.read_csv('P4AI_BT1.csv') 
```

Bài 1: Hiện thị dữ liệu

```python
data
```

Bài 2: Điền các giá trị bị thiếu của dữ liệu theo phương pháp trung bình với các trường có giá trị số và theo phương pháp tần số cao nhất với giá trị rời rạc

```python
df1 = data.copy()
df1
```

```python
df1.isna().sum() 
```

```python
# Chuẩn hóa các giá trị số sử dụng phương pháp trung bình 
df1['sepal.length'].fillna(round(data['sepal.length'].mean()), inplace=True)
df1['sepal.width'].fillna(round(data['sepal.width'].mean()), inplace=True)
df1['petal.length'].fillna(round(data['petal.length'].mean()), inplace=True)
df1['petal.width'].fillna(round(data['petal.width'].mean()), inplace=True)

# Chuẩn hóa các giá trị rời rạc sử dụng phương pháp tần số cao nhất 
df1['variety'].fillna(data['variety'].mode()[0], inplace=True)

df1 

```

Bài 3: Hiển thị các bản ghi có sepal.length > 5 và sepal.width > 3

```python
df2 = df1.copy() 

df2[(df2['sepal.length'] > 5) & (df2['sepal.width'] > 3)]
```

Bài 4: Chuẩn hóa dữ liệu theo min max


![m](https://github.com/UocNTh/python_for_AI/blob/main/Image/min_max_normalisation.jpg?raw=true) 



Trong đó: 

- Xnew    : Giá trị sau khi chuẩn hóa 

- Xi      : Giá trị trước khi chuẩn hóa 

- min(X)  : Giá trị nhỏ nhất của trường dữ liệu 

- max(X)  : Giá trị lớn nhất của trường dữ liệu

```python
df3 = df2.copy() 

def normalized_data(df, p) : 
    p_min = df[p].min() 
    p_max = df[p].max() 

    df[p] = (df[p] - p_min)/(p_max - p_min)
```

```python
normalized_data(df3,'sepal.length')
normalized_data(df3,'sepal.width')
normalized_data(df3,'petal.length')
normalized_data(df3,'petal.width')
df3
```

Bài 5: Liên tục hóa thuộc tính variety


Liên tục hóa một thuộc tính trong AI có thể được hiểu như việc chuyển đổi một thuộc tính có dạng không số hoặc có dạng rời rạc thành dạng số liên tục.

Điều này thường được thực hiện để các thuộc tính đó có thể được sử dụng trong các thuật toán học máy hoặc mô hình AI có yêu cầu dữ liệu số.

```python
df4 = df3.copy() 
df4['variety'].unique() 
```

```python
dummies = pd.get_dummies(df4['variety'])
dummies
```

```python
# Nối các trường dữ liệu 

df4 = pd.concat([df4,dummies], axis=1) 
df4
```

```python
# Xóa bỏ trường variety 

df4.drop('variety', axis=1, inplace=True)
df4
```

Bài 6: Rời rạc hóa thuộc tính sepal.length 

- bins  : là khoảng cần rời rạc
- labels: là nhãn gắn vào các khoảng tương ứng

```python
df5 = df4.copy() 
df5['sepal.length'] = pd.cut(df5['sepal.length'], bins=[-0.1, 0.2, 0.4, 0.6, 0.8, 1], labels=["Very Low", "Low", "Medium", "High", "Very High"])
df5 
```

Bài 7: Lấy mẫu ngẫu nhiên có lặp lại 50% dữ liệu

Cách lấy mẫu khung dữ liệu trong Python Pandas

- Lấy mẫu ngẫu nhiên
- Lấy mẫu với điều kiện 
- Lấy mẫu với tốc độ không đổi

Lấu mẫu ngẫu nhiên

Cho một khung dữ liệu có N hàng, lấy mẫu ngẫu nhiên trích xuất X hàng từ khung dữ liệu

Python Pandas cung cấp hàm sample()

Số lượng mẫu cần thiết có thể được biểu thị theo hai cách khác nhau:

- Chỉ định số hàng ngẫu nhiên chính xác để trích xuất
- Chỉ định phần trăm các hàng ngẫu nhiên để trích xuất. Tỷ lệ phần trăm được biểu thị bằng một số từ 0 đến 1.


```python
df6 = df5.copy() 

sample_data = df6.sample(frac=0.5)

sample_data
```
