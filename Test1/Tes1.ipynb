 Dựa vào tập train (Train_samsung.csv) và điền nhãn cho tập dữ liệu test (Test_samsung_noclass.csv).

```python
import pandas as pd 
data = pd.read_csv('Train_samsung.csv')
data
```

```python
data.isna().sum()
```

Làm sạch dữ liệu 


```python
data.info()
```

```python
# điền dữ liệu liên tục

def fillna_num(data) :
    col_name = data.columns  
    for col in col_name: 
        if data[col].dtype in ['int64', 'float64'] : 
            data[col].fillna(round(data[col].mean()), inplace = True)

def fillna_object(data) : 
    col_name = data.columns  
    for col in col_name: 
        if data[col].dtype == 'object' : 
            data[col].fillna(data[col].mode()[0], inplace = True)

fillna_num(data)
fillna_object(data)

data
```

```python
data['Class'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder 

encoder = LabelEncoder() 

col_name = data.columns   

for col in col_name : 
    if data[col].dtype == 'object' : 
        data[col] = encoder.fit_transform(data[col])

data
```

```python
from sklearn.preprocessing import MinMaxScaler 

scaler = MinMaxScaler() 

data_copy = data.drop('Class', axis=1)

col_name  = data_copy.columns 


normalized_data = scaler.fit_transform(data_copy)

data_copy = pd.DataFrame(normalized_data, columns=col_name)

data = pd.concat([data_copy, data['Class']], axis= 1)

data 
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(
    data.iloc[:, :-1], data.iloc[:, -1], test_size=0.3, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
import pickle

# Huấn luyện mô hình
regressor = LogisticRegression()
regressor.fit(X_train, y_train)

# Dự đoán
predictions = regressor.predict(X_test)

# Đánh giá
accuracy = accuracy_score(y_test, predictions)
f1 = f1_score(y_test, predictions, average="macro")

# In kết quả đánh giá
print("LR classification accuracy:", accuracy)
print("LR f1 score:", f1)

# Lưu mô hình thành tệp tin .pkl
with open('model.pkl', 'wb') as file:
    pickle.dump(regressor, file)
```
